```python
# S02T04 - Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# S02T04 - Carregar o dataset
df = pd.read_csv('../alzheimers_disease_data.csv')
```

```python
# S02T04 - Analise de Outliers
colunas_relevantes = [
    "PatientID", "Age", "Gender", "Ethnicity", "EducationLevel", "BMI", "Smoking", 
    "AlcoholConsumption", "PhysicalActivity", "DietQuality", "MemoryComplaints", 
    "BehavioralProblems", "ADL", "Confusion", "Disorientation", "PersonalityChanges", 
    "DifficultyCompletingTasks", "Forgetfulness", "Diagnosis", "DoctorInCharge",
    "SystolicBP", "DiastolicBP", "CholesterolTotal", "CholesterolLDL", 
    "CholesterolHDL", "CholesterolTriglycerides"
]

for coluna in colunas_relevantes:
    if coluna in df.columns:
        if df[coluna].dropna().nunique() > 3:
            plt.figure(figsize=(8, 4))
            plt.boxplot(df[coluna].dropna(), vert=False)
            plt.title(f'Boxplot de {coluna}')
            plt.xlabel(coluna)
            plt.show()
```

```python
# S02T04 - Conclusao

print('Aps a anlise, foi identificado que h alguns falsos outliers nas colunas Ethnicity e EducationLevel, no sendo necessrio realizar qualquer tratamento.')
```
